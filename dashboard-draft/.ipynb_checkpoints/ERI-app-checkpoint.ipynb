```python
# LOAD
import os
import pandas as pd
import numpy as np
import plotly.express as px
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
```

```python
# DATA, VARIABLES
#external_stylesheets = ['https://codepen.io/chriddyp/pen/bWLwgP.css']
#mytitle='ERI projects & publications (2009 - 2019) clustered by topic'
#tabtitle='ERI Research Map'
#githublink='https://github.com/saralafia/ERI-dashboard'

df = pd.read_csv('70-document-topic-viz.csv', index_col=0)
#df.head()

reshaped = \
(df.set_index(df.columns.drop('authors',1).tolist())
   .authors.str.split('; ', expand=True) #split PI column by ;
   .stack()
   .reset_index()
   .rename(columns={0:'authors'})
   .loc[:, df.columns]
)

reshaped[['Last_Name', 'First_Name']] = reshaped.authors.str.split(",",expand=True)
reshaped['Last_Name'] = reshaped['Last_Name'].str.strip()
reshaped['First_Name'] = reshaped['First_Name'].str.rstrip('ABCDEFGHIJKLMNOPQRSTUVWXYZ.')
reshaped['First_Name'] = reshaped['First_Name'].str.strip()
reshaped['Name'] = reshaped['First_Name'] + " " + reshaped['Last_Name']
reshaped.loc[reshaped['Name'].isnull(),'Name'] = reshaped['authors'] 

reshaped_dropped = reshaped.drop(columns=['authors', 'Last_Name','First_Name'])
#reshaped_dropped.dtypes

result = reshaped_dropped.merge(df)
result['Name'] = reshaped['Name'].str.strip()
# result.head(20)
# result.to_csv('merged.csv')

ERI_pis = pd.read_csv('ERI-people-2009-2019.csv', index_col=0)
#pis_list = ERI_pis['Name'].to_list()

pi_result = result.merge(ERI_pis)
pi_result.tail(20)
#pi_result.dtypes
pi_result.to_csv('data-70.csv')
#print(len(pi_result))

# pi_list = pi_result['Name'].unique().tolist()
# pi_list[0]
#print(len(pi_list))

# topic_0 = pi_result['topic0'].sum()
# topic_1 = pi_result['topic1'].sum()
# topic_2 = pi_result['topic2'].sum()
# topic_3 = pi_result['topic3'].sum()
# topic_4 = pi_result['topic4'].sum()
# topic_5 = pi_result['topic5'].sum()
# topic_6 = pi_result['topic6'].sum()
# topic_7 = pi_result['topic7'].sum()
# topic_8 = pi_result['topic8'].sum()

# bar = pd.DataFrame()
# bar['number']  = pi_result['main_topic'].sort_values().unique()
# bar['topic']  = pi_result['main_label'].unique()
#bar['weight'] = pi_result['main_label'].unique()
```
